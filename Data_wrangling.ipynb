## Concating Pandas data frame

```python
data_0 = {'id':[1,2,3],'Name':['Ramesh','Suresh','Mahesh'],'City':['Hyd','Mum','Kolkata']}
data_1 = {'id':[4,5,6],'Name':['Sachin','Dhoni','Virat'],'City':['Mum','Jarkhand','Delhi']}
print(pd.DataFrame(data_0))
print(pd.DataFrame(data_1))
df_0 = pd.DataFrame(data_0)
df_1 = pd.DataFrame(data_1)

```

```python
pd.concat([df_0,df_1],axis=0,ignore_index=True)
df = pd.concat([df_0,df_1],axis=0,ignore_index=True)
df
```

```python
## alternatively append new row to a dataframe

new_row = pd.Series(['7','Rohit','Ben'],index=['id','Name','City'])
new_row
```

```python
df.append(new_row,ignore_index=True)
```

### Merging DataFrame

```python
cric_ratings = {'Rank':[1,2,3,4,5],'Cricketer':['Virat','Babar','Rohit','Smith','Williams']}
countries = {'Rank':[1,2,7,3],'Country':['India','Pakistan','New Zealand','India']}

cric_ratings_df = pd.DataFrame(cric_ratings)
countries_df = pd.DataFrame(countries)

pd.merge(cric_ratings_df,countries_df,on='Rank',how='right')
```
