```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
```

```python
domain=np.linspace(-2,2,1000)
plt.plot(domain,norm.pdf(domain,0,1))
plt.title('Standard Deviation')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
data = [186, 176, 158, 180, 186, 168, 168, 164, 178, 170, 189, 195, 172,
     187, 180, 186, 185, 168, 179, 178, 183, 179, 170, 175, 186, 159,
     161, 178, 175, 185, 175, 162, 173, 172, 177, 175, 172, 177, 180]
```

```python
data.sort()
```

```python
data_mean=np.mean(data)
data_std=np.std(data)
print(data_mean)
print(data_std)
```

```python
pdf=norm.pdf(data,data_mean,data_std)
plt.plot(data,pdf)
plt.title('Normal distribution example')
plt.xlabel('x axis value')
plt.ylabel('y axis')
plt.show()
```
